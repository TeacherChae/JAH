```python
# 📝 과제: 몬테카를로 방법으로 원주율 추정하기

# 1. 문제 배경
# 우리는 흔히 수학적 상수를 **계산식**으로 배우지만, 컴퓨터를 활용하면 **확률적 방법**으로도 근사할 수 있습니다.  
# 그 대표적인 방법이 **몬테카를로(Monte Carlo) 시뮬레이션**입니다.  

# 몬테카를로 방법은 무작위로 점을 많이 찍고, 그 결과를 통계적으로 처리해서 원하는 값을 추정하는 방식입니다.  
# 이번 과제에서는 몬테카를로 방법을 사용하여 **원의 넓이**를 근사하고, 이를 통해 **원주율 π를 추정**해보겠습니다.  
```

```python
from random import random # random 라이브러리의 random()함수는 0~1사이의 실수를 호출마다 무작위로 하나씩 반환합니다.
import math # 사칙 연산보다 구체적인 대수적 연산이 필요할 때 사용하는 파이썬 내장 라이브러리입니다.


count_limit = 10  # 임의값이므로 수정하셔도 됩니다.
# 1. 반복문을 돌기 전에 어떤 값을 미리 초기화해야할까요?
total_count = 0
inside_count = 0

for i in range(count_limit):
    x = random()
    y = random()
    total_count += 1
    if x**2 + y**2 <= 1:
        inside_count += 1
    # 2. x, y는 0과 1사이 무작위 실수입니다. 두 값을 좌표로 하는 점이 반지름이 1인 원 안에 있음을 판별하려면 어떤 조건문이 필요할까요?
    # 3. 전체 점 개수와 원 안에 있는 점 개수의 관계는 무엇을 의미할까요?
    # 4. iteration을 돌면서 특정 값에 근사하고 있는지 확인해봅시다.
    # 5. count_limit을 늘려가면서 특정 값에 더 잘 근사하고 있는지 확인해봅시다.
ratio = inside_count / total_count

print(ratio)
print(math.pi / 4)  # math.pi는 원주율 π를 소수점 16자리까지 나타냅니다.
```
