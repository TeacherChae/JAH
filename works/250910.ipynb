```python
# 1. 숫자가 양수인지, 음수인지, 0인지 판별하기

def check_positive(n):
    if n > 0:
        return "positive"
    elif n == 0:
        return "zero"
    else:
        return "negative"
x = 1

print(check_positive(x))
```

```python
# 2. 입력한 수가 짝수인지 홀수인지 판별하기
n = int(input())
if n % 2 == 0:
    print("even number")
else:
    print("odd number")
print(n % 2)
```

```python
# 3. 세 과목 점수를 입력받아 평균이 60점 이상이면 합격, 아니면 불합격 출력하기
scores = list(map(int, input().split()))
average = sum(scores) / len(scores)
print(average)
if average >= 60:
    print('pass')
else:
    print('fail')

```

```python
# 4. 좌표 (x,y)가 제1사분면, 제2사분면, 제3사분면, 제4사분면 중 어디에 속하는지 판별하기
x, y = 3, -3
def evalute_quadrant(x, y):
    if x > 0 and y > 0:
        print("제1사분면")
    elif x < 0 and y > 0:
        print("제2사분면")
    elif x < 0 and y < 0:
        print("제3사분면")
    elif x > 0 and y < 0:
        print("제4사분면")
    elif x == 0 and y != 0:
        print("점이 y축 위에 위치")
    elif x != 0 and y == 0:
        print("점이 x축 위에 위치")
    else:
        print("원점")

evalute_quadrant(x, y)
```

```python
# 5. 점 (x,y)가 원점 중심, 반지름 1 원 내부/경계/외부에 있는지 판별하기
x, y = map(float, input().split())

if x**2 + y**2 < 1:
    print("원 내부")
elif x**2 + y**2 == 1:
    print("원 경계")
else:
    print("원 외부")
```

```python
# 2. 1부터 n까지 합 구하기 (for 문)
n = int(input())
m = 0
for i in range(1, n+1):
    m += i
print(m)

# 2-1.
n = int(input())
m = sum(range(1, n+1))
print(m)
```
